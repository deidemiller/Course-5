```python
#basic imports
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt 
```

```python
#import data
rawData = pd.read_csv('default of credit card clients.csv', header=1)
rawData.head()
```

```python
classData = rawData
```

```python
#remove unnecessary and problematic columns
classData = classData.drop(['SEX', 'ID'], axis = 1)
```

```python
classData.head()
```

```python
#check correlation nothing seems too correlatated with default payments
corrMat = classData.corr()
print(corrMat)
```

```python
#remove negative balances on bill amount 6
classData = classData[classData['BILL_AMT6'] > 0]
```

```python
classData.describe()
```

```python
covMat = classData.cov()
print(covMat)
```

```python
depVar = classData['default payment next month']
```

```python
features = classData.iloc[:,0:22]
print("Summary of Features")
features.head()
```

```python
features.describe()
```

```python
X_train = (features[: 10000])
X_train.head()
X_train.describe()
```

```python
y_train = depVar[: 10000]
y_train_count = len(y_train.index)
print('The number of observations in the Y training set are:',str(y_train_count))
y_train.head()
```

```python
#Testing Set (X Testing)
X_test = features[-1000:]
X_test_count = len(X_test.index)
print('The number of observations in the feature testing set is:',str(X_test_count))
print(X_test.head())
```

```python
#Ground Truth (y_test) 
y_test = depVar[-1000:]
y_test_count = len(y_test.index)
print('The number of observations in the Y training set are:',str(y_test_count))
y_test.head()
```

```python
#split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train)
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rfModel = RandomForestClassifier(max_depth=2, random_state=0)
rfModel.fit(X_train, y_train)
rfModel.score(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
print(cross_val_score(rfModel, X_train, y_train)) 
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbcModel = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
        max_depth=1, random_state=0).fit(X_train, y_train)
gbcModel.score(X_train, y_train)
```

```python
print(cross_val_score(gbcModel, X_train, y_train)) 
```

```python
from sklearn import tree
treeModel = tree.DecisionTreeClassifier(random_state=0, max_depth=2)
treeModel = treeModel.fit(X_train, y_train)
treeModel.score(X_train, y_train)
```

```python
print(cross_val_score(treeModel, X_train, y_train)) 
```

```python
gbcPredict = gbcModel.predict(X_test)
rfPredict = rfModel.predict(X_test)
treePredict = treeModel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
gbcReport = classification_report(gbcPredict, y_test)
print(gbcReport)
```

```python
treeReport = classification_report(treePredict, y_test)
print(treeReport)
```

```python
rfReport = classification_report(rfPredict, y_test)
print(rfReport)
```
